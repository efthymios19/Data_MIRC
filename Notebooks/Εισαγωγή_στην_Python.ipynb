Μετά το όνομα της συνάρτησης ακολουθούν σε παρένθεση οι ορισμένες μεταβλητές, οι οποίες αποτελούν τα απαραίτητα δεδομένα για την εκτέλεση των εντολών μέσα στην συνάρτηση. ### Ροές ελέγχου, for loops

Η βασική ιδέα πίσω από τις ροές ελέγχου και ειδικότερα τα **for loops** είναι η απλούστευση ενεργειών μέσα στο σενάριο (script) του κώδικα. Σκεφτείτε ότι έχετε μια λίστα με συμβολοσειρές για αντικείμενα, τα οποία θέλετε να τυπώσετε ένα προς ένα. Θα πρέπει να χρησιμοποιήσετε τον κατάλληλο δείκτη για κάθε ένα αντικείμενο της λίστας:

```
list=['Neoxwraki', 'Grammenitsa', 'Louros', 'Aracthos']

list[0]
list[1]
list[2]
list[3]
```
Ο συγκεκριμένος τρόπος, όμως δεν είναι αποδοτικός, ειδικά για μεγάλες λίστες, και γι' αυτό μπορούμε να χρησιμοποιήσουμε τις ροές ελέγχου.

H γενική μορφή των for loops στην python είναι η εξής:

```
for variable in collection:
    do things with variable

```

1.   Το keyword variable (μεταβλητή) μπορεί να είναι οποιδήποτε όνομα της αρεσκείας σας, το οποίο να μην έχει επαναχρησιμοποιηθεί.
2.   Η for loop τελειώνει με (:)
3. H εσοχή ( identation) είναι συνήθως 4 spaces.





````python
# Έστω ότι προσδιορίζουμε την παραπάνω λίστα με τις συμβολοσειρές:

list_stations=['Neoxwraki', 'Grammenitsa', 'Louros', 'Aracthos']

# Για να τυπώσουμε όλα τα αντικείμενα της λίστας μπορούμε να χρησιμοποιήσουμε μία for-loop:

for i in list_stations:
  print(i) # η for-loop διατρέχει όλα τα αντικείμενα της λίστας
````

````python
# H εξίσωση range()[help(range)]

#range(start,stop, step) ή range(stop) ξεκινάει από το 0 (default) και τελειώνει
#ένα αριθμό πριν αυτόν που έχουμε προσδιορίσει

list_range=range(8)
print(list_range)

#Για να δημιουργήσετε μια λίστα με την εξίσωση range():

new_list=[*range(8)]
print(new_list)
````

````python
# H εξίσωση range σε for-loop:

for i in range(10):
  print(i)
````

Διατρέχοντας τις λίστες χρησιμοποιώντας τις τιμές των δεικτών τους

Χρησιμοποιώντας την συνάρτηση len() μπορούμε να προσδιορίσουμε τον αριθμό των στοιχέιων μιας λίστας, μπορούμε να κάνουμε πιο ευέλικτες τις for-loops. Μπορούμε χρησιμοποιώντας σε συνδυασμό τις συναρτήσεις len() και range() να τυπώσουμε τα ονόματα της λίστας των σταθμών μαζί με το δείκτη τους.


````python
for i in range(len(list_stations)):
  print(f"The station {list_stations[i]} is at index {i} in the list_stations")
````

H πρόσθεση Gauss και το παράδοξο της επαναληψιμότητας

Αν σας έλεγα να προσθέσετε όλους τους αριθμούς από το 1 έως το 100 χρησιμοποιώντας την μέθοδο των διαδοχικών προσθέσεων θα δαπανούσατε αρκετό χρόνο ώστε να φτάσετε στο αποτέλεσμα. Όταν το συγκεκριμένο πρόβλημα τέθηκε στον διάσημο μαθηματικό Gauss, όταν αυτός ήταν μαθητής ακόμα, πολύ γρήγορα έδωσε την σωστή απάντηση 5050.

Ο Gauss αντιλήφθηκε πως η πρόσθεση κάθε ζεύγους αριθμών από την αρχή έως το τέλος της δωθήσας λίστα (0 - 100) άθροιζε 101:

image.png

Κατά αυτόν τον τρόπο για το άθροισμα διαδοχικών αριθμών μπορεί να αποφευχθεί η επαναληπτική μέθοδος και να χρησιμοποιηθεί ο τύπος της πρόσθεσης Gauss:

$$Total_{sum}= number-of-pairs * sum-of-each-pair$$

$$number-of-pairs= \frac{\max(list-of-numbers)}{2}$$

$$sum-of-each-pair= \max(list-of-numbers) + 1$$

Αρκεί να προσδιορίσουμε τον αριθμό των ζευγών και το άθροισμα τους. Για παράδειγμα για τους αριθμούς από το ένα ώς το 10 τα ζεύγη είναι 5 και το άθροισμα τους 11. Διαιρούμε τον μέγιστο αριθμό της λίστας/2 (για να βρούμε τα ζεύγη) και πολλαπλασιάζουμε με το κοινό άθροισμα τους.

````python
# Υπολογισμός αθροίσματος 100 διαδοχικών αριθμών με τη μέθοδο της διαδοχικής πρόσθεσης
value=0
List_mat=[*range(0,101,1)]
for i in List_mat:
  value+=i
print(value)
````

````python
# Υπολογισμός αθροίσματος 100 διαδοχικών αριθμών με τη μέθοδο της πρόσθεσης Gauss

def GaussianAdd (mat):
  return mat[-1]/2 * (mat[-1]+1)

GaussianAdd(List_mat)
````

#### Εξερευνoντάς υδρολογικά δεδομένα με την βιβλιοθήκη pandas

**Τι είναι ένα API;**

Ένα API, ή διεπαφή προγραμματισμού εφαρμογών (application programming interface), χρησιμοποιείται για τη μεταβίβαση δεδομένων μεταξύ εφαρμογών λογισμικού με έναν τυποποιημένο τρόπο. Πολλές υπηρεσίες προσφέρουν δημόσια API που επιτρέπουν σε οποιονδήποτε να στέλνει και να λαμβάνει περιεχόμενο από την υπηρεσία.

Χρήση API για τη λήψη υδρολογικών δεδομένων από το δίκτυο [e-ΠΥΡΡΟΣ](http://e-pyrros.gr/field-lab-gr/)

````python
import requests
import pandas as pd
import ast
import numpy as np
````

````python
# Κάνουμε κλήση στο API του e-ΠΥΡΡΟΣ ώστε να πάρουμε τα δεδομένα σε μορφή json

response_API=requests.get('https://df.e-pyrros.gr/api/v2/lora/_table/event_up?fields=time,device_name,object,rx_info,&filter=device_name=LSE01_1&limit=100&order=time%20DESC&api_key=d11076d30cc60b2c3d75383be92d6c9805774ca1a63837c050e7fd2e07a890ec').json()

response_API
````

````python
# Eισαγωγή του json αρχείου σε μία pandas Dataframe

df_api=pd.json_normalize(response_API, 'resource')

df_api
````

````python
df_api["object"] #επιλέγουμε συγκεκριμένη στήλη της dataframe
````

````python
# H συνάρτηση expand_str_dict_to_columns λαμβάνει ένα dictionary και
# διαχωρίζει τις επιμέρους του στήλες

def expand_str_dict_to_columns(row):
    row = ast.literal_eval(row) #η συνάρτηση literal_eval της βιβλιοθήκης ast μετατρέπει το nested dictionary σε απλό dictionary
    for key, value in row.items(): #η for-loop διατρέχει όλα τις τιμές του dictionary που λήφθηκαν στο προηγούμενο βήμα
        row[key] = value # αντιστοιχεί κάθε τιμή με το αντίστοιχο κλειδί της
    return row
````

````python
# Με την συνάρτηση apply() της pandas εφαρμόζουμε την συνάρτηση στην στήλη
# object της df_api dataframe και έπειτα κάθε στοιχείου του dictionary
# μετατρέπεται σε στήλη της νέας dataframe

df_api_ = df_api['object'].apply(expand_str_dict_to_columns).apply(pd.Series)
````

````python
# Δημιουργούμε μια νέα dataframe ως συνένωση των δύο προηγούμενων
# καθώς θέλουμε ναα διατηρήσουμε τη στήλη με τις χρονικές τιμές των μετρήσεων
df_api=pd.concat([df_api,df_api_], axis=1)
````

````python
# Απορρίπτουμε τις στήλες που δεν χρειαζόμαστε από την df_api dataframe

df_api=df_api.drop(['object', 'rx_info'], axis=1)

df_api
````

````python
type(df_api["temp_SOIL"][0])
````

````python
# Μετατρέπουμε τις τιμές των στηλών της εδαφικής θερμοκρασίας, της εδαφικής
# υγρασίας και της αγωγιμότητας εδάφους από συμβολοσειρές σε αριθμούς κινητής
# υποδιαστολής

df_api['temp_SOIL']=df_api['temp_SOIL'].astype(float)
df_api['water_SOIL']=df_api['water_SOIL'].astype(float)
df_api['conduct_SOIL']=df_api['conduct_SOIL'].astype(float)
````

````python
# Μετατρέπουμε τους αριθμητικούς δείκτες της dataframe σε ημερολογιακούς

df_api['time'] = pd.to_datetime(df_api['time'])
df_api.set_index('time',inplace=True)

df_api
````

#### Αναδιαμόρφωση του χρονικού βήματος της df_api dataframe

Τα δεδομένα που συλλέγουμε από αισθητήρες πεδίου μας αποδιδούν τιμές σε πυκνό χρονικό διάστημα (10 min, 20min). Για την μετέπειτα επεξέργασια αυτών των δεδομένων μπορούμε να πυκνώσουμε ή να μειώσουμε το χρονικό βήμα των χρονοσειρών.

[resample()](https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.resample.html)


````python
df_api_daily = df_api.resample('D').agg({'temp_SOIL': np.mean, 'water_SOIL': np.mean, 'conduct_SOIL': np.mean})

df_api_daily
````
